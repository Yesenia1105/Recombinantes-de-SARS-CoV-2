Función que recibe el DataFrame y arroja la cantidad de "-" que aparecieron en el genoma de referencia, al momento de alinearlo con el resto de los genomas que estamos estudiando.

```python
def DatosNulos(DF):
    St=DF.iloc[0,0]
    i=0
    while St[i]=='-': i+=1
    return(i)
```

Luego, la siguiente función recibe 
* m y n que identifican los renglones 2n y 2m, es decir los genomas ordenados en las posiciones m y n.
* D que es el DataFrame con el que estamos trabajando, y
* k que denota el k-ésimo genoma a comparar.

Dicha función arroja una tabla con las posiciones donde los k-ésimos genes difieren entre el genoma n y el genoma m. Además, dice cuáles son las entradas en cada uno de ellos.


```python
def dif1(m,n,D,k):
    Dif = pd.DataFrame(columns=('Posicion', 'String1', 'String2'))
    DifN = pd.DataFrame(columns=('Posicion', 'String1', 'String2'))
    for i in range(Ranges[k][0], Ranges[k][1]):
        if D.iloc[2*m,0][i]==D1.iloc[2*n,0][i]:
            continue
        elif D.iloc[2*m,0][i]=='N' or D.iloc[2*n,0][i]=='N':
            DifN.loc[len(DifN)] = (i, D.iloc[2*n,0][i], D1.iloc[2*m,0][i])        
        else: 
            Dif.loc[len(Dif)] = (i, D.iloc[2*n,0][i], D.iloc[2*m,0][i])
    return(Dif, DifN)
```

Ésta es una función que me arroja la matriz de distancias entre todos los genomas que están en el DataFrame alineado

- Recibe un dataframe DF y un m que es la cantidad de genomas a comparar


```python
def M_aco(DF):
    m=int(np.ceil(len(DF)/2))
    #i=DatosNulos(DF)
    M_dist_array = np.zeros(shape=(m,m))
    M_dist=pd.DataFrame(M_dist_array)

    for k in range(m):  
        for j in range(k):
            cont=0
            for i in range(266,29533):
                if DF.iloc[2*k,0][i]==DF.iloc[2*j,0][i]:
                    continue
                elif DF.iloc[2*k,0][i]=='N' or DF.iloc[2*j,0][i]=='N':
                    continue                 
                else:
                    cont=cont+1
            M_dist.iloc[k,j]=cont
            M_dist.iloc[j,k]=cont        
    return(M_dist)
```
